```python
# Dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
# Read in CSV
crypto_df = pd.read_csv("Resources/crypto_data.csv")
crypto_df
```

```python
# Clean Crypto Data
## Eliminate columns with null value
drop_crypto_df = crypto_df.dropna(how='any')

## Filter for those currently being mined and trading
filtered_crypto_df = drop_crypto_df.loc[(drop_crypto_df['IsTrading'] == True) & (drop_crypto_df['TotalCoinsMined'] > 0), :]
filtered_crypto_df

## Drop the IsTrading and CoinName columns
clean_crypto_df = filtered_crypto_df.drop(columns = ['IsTrading', 'CoinName', 'Unnamed: 0'])
clean_crypto_df
```

```python
# Prepare data for machine learning
## Convert objects into numerics
crypto_dummies = pd.get_dummies(clean_crypto_df)
crypto_dummies
```

```python
## Scale Data
scaler = StandardScaler().fit(crypto_dummies)
crypto_scaled_df = scaler.transform(crypto_dummies)
crypto_scaled_df
```

```python
# Reduce number of inputs
## PCA model
pca = PCA(n_components=0.9)

crypto_pca = pca.fit_transform(crypto_scaled_df)

# Transform PCA data to a DataFrame
crypto_pca_df = pd.DataFrame(
    data=crypto_pca)
crypto_pca_df.head()
```

```python
# Use t-SNE to reduce the features of the model
## Intialize t-SNE
tsne = TSNE(learning_rate = 100)

## Reduce dimensions
tsne_crypto = tsne.fit_transform(crypto_pca_df)

## Check Shape
tsne_crypto
```

```python
# Plot t-SNE

## The first column of transformed features
crypto_pca_df['x'] = tsne_crypto[:,0]

## The second column of transformed features
crypto_pca_df['y'] = tsne_crypto[:,1]

## Plot
plt.scatter(crypto_pca_df['x'], crypto_pca_df['y'])
plt.show()
```

```python
# Create a t-SNE data frame
crypto_tsne_df = pd.DataFrame(
    data=tsne_crypto, columns=["principal component 1", "principal component 2"]
)
crypto_tsne_df.head()
```

```python
# Clustering and Analysis with k-means
## Finding best k
inertia = []
k = list(range(1, 11))

# Calculating inertia
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_tsne_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve
elbow_curve = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_curve)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Predicting clusters with k=4

# Initialize the k-means model
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(crypto_tsne_df)

# Predict clusters
predictions = model.predict(crypto_tsne_df)

# Predicted classification column
crypto_tsne_df["class"] = model.labels_
crypto_tsne_df
```
